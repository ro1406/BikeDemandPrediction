Attempt 1: Remove outliers from Bike Rent too!

```python
print(len(data))
for col in copy.columns[:9]:
    ind = copy[(copy[col]>=3) | (copy[col]<=-3)].index
    copy.drop(ind,inplace=True)
    data.drop(ind,inplace=True)
print(len(data))
```
